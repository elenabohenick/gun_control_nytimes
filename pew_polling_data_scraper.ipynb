```python
import requests
from bs4 import BeautifulSoup
import string
import re
import pandas as pd
```

```python
url = 'http://www.people-press.org/2017/06/22/public-views-about-guns/#all'

response=requests.get(url)
page=response.text

soup=BeautifulSoup(page,"html5lib")
```

```python
# tables=soup.find_all(class_='region-holder datatable')
```

### Grab Overall table

```python
table = soup.find_all(class_='total-holder datatable')
df = pd.read_html(str(table))
df = df[0]
df
```

```python
df.rename(columns={'Unnamed: 0':'issue'},inplace=True)
```

```python
df['issue'] = [x.split(' ')[1] for x in df['issue']]
df.set_index('issue',inplace=True)
```

```python
df.head()
```

```python
df = df.T
```

```python
df = pd.DataFrame({'date':df.index,'rights_poll':df.values[:,0], 'control_poll':df.values[:,1]})
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
cols = ['date','rights_poll','control_poll']

df = df.loc[:, cols]
df
```

```python
df.to_csv('../data/pew_overall.csv')
```

### By Region

```python
table = soup.find_all(class_='region-holder datatable')
df = pd.read_html(str(table))
df = df[0]
df
```

```python
df.rename(columns={'Unnamed: 0':'Region_issue'},inplace=True)
```

```python
df['region'] = [x.split(' ')[0] for x in df['Region_issue']]
df['issue'] = [x.split(' ')[1] for x in df['Region_issue']]
```

```python
df.drop('Region_issue',axis=1,inplace=True)
```

```python
cols = ['region', 'issue','12/5/93', '5/16/99', '6/14/99', '3/19/00', '4/16/00', '5/6/00',
       '6/8/03', '2/16/04', '4/22/07', '11/26/07', '4/27/08', '4/21/09',
       '3/14/10', '9/6/10', '1/16/11', '3/1/11', '10/4/11', '4/15/12',
       '7/29/12', '12/19/12', '1/13/13', '2/18/13']
df = df.loc[:, cols]
df
```

```python
df.set_index('region',inplace=True)
```

```python
df_rights = df[df.issue=='rights'].copy()
df_control = df[df.issue=='control'].copy()
```

```python
df_rights.drop(labels='issue',axis=1,inplace=True)
df_control.drop(labels='issue',axis=1,inplace=True)
```

```python
df_rights = df_rights.T
df_control = df_control.T
```

```python
df_rights.reset_index(level=0, inplace=True)
df_control.reset_index(level=0, inplace=True)
```

```python
df_rights.rename(columns={'index':'date'},inplace=True)
df_rights['date'] = pd.to_datetime(df_rights['date'])

df_control.rename(columns={'index':'date'},inplace=True)
df_control['date'] = pd.to_datetime(df_control['date'])
```
